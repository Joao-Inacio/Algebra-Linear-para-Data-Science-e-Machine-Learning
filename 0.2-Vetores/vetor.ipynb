## Regularização

```python
coef_x1 = 2.5
coef_x2 = -1.8
coef_x3 = 0.05
intercept = 3.0
```

```python
np.random.seed(10)
```

```python
x1 = np.random.uniform(0, 10, size=10)
x2 = np.random.uniform(0, 10, size=10)
x3 = np.random.uniform(0, 10, size=10)
```

```python
x1, x2, x3
```

```python
y = coef_x1 * x1 + coef_x2 * x2 + coef_x3 * x3 + intercept
```

```python
X = np.column_stack((x1, x2, x3))
```

```python
X
```

```python
X.shape
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.linear_model import Lasso
```

```python
model_lasso = Lasso()
```

```python
model_lasso.fit(X, y)
```

```python
model_lasso.coef_
```

```python
model.intercept_
```

```python
from sklearn.linear_model import Ridge
```

```python
model_ridge = Ridge()
```

```python
model_ridge.fit(X, y)
```

```python
model_ridge.coef_
```

```python
model_ridge.intercept_
```
