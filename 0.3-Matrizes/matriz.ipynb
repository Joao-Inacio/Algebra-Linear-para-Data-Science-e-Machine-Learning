## Matriz simétrica

```python
sym_matrix = np.array([
    [1, 3, 2, 5],
    [3, 2, 8, 1],
    [2, 8, 6, 7],
    [5, 1, 7, 4],
])
```

```python
scipy.linalg.issymmetric(sym_matrix)
```

```python
np.all(sym_matrix == sym_matrix.T)
```

## Matriz diagonal

```python
diag_matrix = np.array([
    [1, 0, 0, 0],
    [0, 2, 0, 0],
    [0, 0, 6, 0],
    [0, 0, 0, 4]
])
```

```python
diagonal_elements = np.diag(diag_matrix)
diagonal_elements
```

```python
new_diag_matrix = np.diag(diagonal_elements)
new_diag_matrix
```

```python
np.all(diag_matrix == new_diag_matrix)
```

```python
diagonal_elements_reciprocal = np.reciprocal(diagonal_elements.astype(float))
diagonal_elements_reciprocal
```

```python
diag_matrix_inverse = np.linalg.inv(diag_matrix)
diag_matrix_inverse
```

```python
np.all(diagonal_elements_reciprocal == np.diag(diag_matrix_inverse))
```
