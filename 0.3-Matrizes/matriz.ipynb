## Matriz simétrica

```python
sym_matrix = np.array([
    [1, 3, 2, 5],
    [3, 2, 8, 1],
    [2, 8, 6, 7],
    [5, 1, 7, 4],
])
```

```python
scipy.linalg.issymmetric(sym_matrix)
```

```python
np.all(sym_matrix == sym_matrix.T)
```
